```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['Loan_ID'] = lb.fit_transform(df['Loan_ID'])
df['Gender'] = lb.fit_transform(df['Gender'])
df['Married'] = lb.fit_transform(df['Married'])
df['Dependents'] = lb.fit_transform(df['Dependents'])
df['Education'] = lb.fit_transform(df['Education'])
df['Self_Employed'] = lb.fit_transform(df['Self_Employed'])
df['Property_Area'] = lb.fit_transform(df['Property_Area'])
df['Loan_Amount_Term'] = lb.fit_transform(df['Loan_Amount_Term'])
df['Credit_History'] = lb.fit_transform(df['Credit_History'])
df['LoanAmount(in K)'] = lb.fit_transform(df['LoanAmount(in K)'])
df.dtypes
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,1:-4]
y = df.iloc[:,-4]
print(x.shape,y.shape)
print(type(x),type(y))
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.87)
print(x_tr.shape)
print(x_te.shape)
print(y_tr.shape)
print(y_te.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
def apply_model(model):
    model.fit(x_tr,y_tr)
    print('Accuracy',model.score(x_te,y_te))
    ypred = model.predict(x_te)
    print('Predicted values\n',ypred)
    cm = confusion_matrix(y_te,ypred)
    print('Confusion Matrix\n',cm)
    sns.heatmap(cm,annot=True)
    plt.show()
    print('Classification report\n',classification_report(y_te,ypred))
```

```python
# 1) Decision Tree
m1 = DecisionTreeClassifier(criterion='gini',max_depth=1,min_samples_split=3)
apply_model(m1)
```

```python
# 2) LogReg
m2 = LogisticRegression()
apply_model(m2)
```

```python
# 3) Random Forest
m3 = RandomForestClassifier(n_estimators=70,criterion='gini',max_depth=5,min_samples_split=10)
apply_model(m3)
```

```python
# 4) SVC
m4 = SVC(kernel='linear',C=1)
apply_model(m4)
```

```python
# 5) KNN
m5 = KNeighborsClassifier(n_neighbors=27)
apply_model(m5)
```
